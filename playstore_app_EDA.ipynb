```python
# Convert 'Size' column to numerical format
def convert_size_to_mb(size_str):
    if pd.isna(size_str):
        return None
    size_str = size_str.replace(',', '')  # Remove commas
    if 'M' in size_str:
        return float(size_str.replace('M', ''))
    elif 'K' in size_str:
        return float(size_str.replace('K', '')) / 1000  # Convert kilobytes to megabytes
    else:
        return None

df['Size'] = df['Size'].apply(convert_size_to_mb)
```

```python
df['Type'] = np.where(df['Free'], 'Free', 'Paid')
df.drop(['Free'], axis=1, inplace=True)
```

```python
# Handling 'Minimum Android' column (keeping only numeric part)
df['Minimum Android'] = df['Minimum Android'].str.extract('(\d+\.\d+)')
```

```python
# Handling 'Content Rating' categories
df['Content Rating'] = df['Content Rating'].replace({
    'Mature 17+': 'Adults',
    'Everyone 10+': 'Teen',
    'Unrated': 'Everyone',
    'Adults only 18+': 'Adults'
})
```

```python
df['Content Rating'].unique()
```

```python
# Handling 'Rating Type' categories based on 'Rating Count'
df['Rating Type'] = 'NoRatingProvided'
df.loc[(df['Rating Count'] > 0) & (df['Rating Count'] <= 10000.0), 'Rating Type'] = 'Less than 10K'
df.loc[(df['Rating Count'] > 10000) & (df['Rating Count'] <= 500000.0), 'Rating Type'] = 'Between 10K and 500K'
df.loc[(df['Rating Count'] > 500000) & (df['Rating Count'] <= 138557570.0), 'Rating Type'] = 'More than 500K'
```

```python
top_categories = df.Category.value_counts().reset_index().rename(columns={'Category':'Count','index':'Category'})
top_categories
```

```python
categories_installed = df.groupby(['Category'])[['Installs']].sum()
categories_installed
```

```python
top_category_installation = pd.merge(top_categories, categories_installed, on='Category')
top_category_installation.head()
```

```python
top_10_Categories_installs = top_category_installation.head(10).sort_values(by = ['Installs'],ascending = False)
```

```python
# Question 1: Top 10 Categories Installed

plt.figure(figsize=(14,7))
plt.xticks(rotation=60)
plt.xlabel("Category")
plt.ylabel("Number of applications")
plt.title("Top 10 Installed Categories")
sns.barplot(x = top_10_Categories_installs.Category,y= top_10_Categories_installs.Installs, palette= 'tab10')
     
```

```python
plt.figure(figsize=(14,7))
plt.xticks(rotation=90)
plt.xlabel("Category")
plt.ylabel("Number of applications")
plt.title("Total Categories and count of Applications in each category ")
sns.barplot(x = top_category_installation.Category, y = top_category_installation.Count, palette= 'tab20b')
```

```python
# Question 3: Highest Rated Category
average_ratings_by_category = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=average_ratings_by_category.values, y=average_ratings_by_category.index, palette='tab10')
plt.title('Average Ratings by Category')
plt.xlabel('Average Rating')
plt.ylabel('Categories')
plt.show()
```

```python
# Question 4: Category with the Highest Paid and Free Apps
category_paid_free = df.groupby(['Category', 'Type']).size().unstack().fillna(0)
category_paid_free.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20b')
plt.title('Distribution of Paid and Free Apps in Each Category')
plt.xlabel('Categories')
plt.ylabel('Number of Apps')
plt.show()
```

```python
# Question 5: Impact of Application Size on Installations
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Size', y='Installs', data=df, alpha=0.5, color = 'green')
plt.title('Impact of Application Size on Installations')
plt.xlabel('Size (in MB)')
plt.ylabel('Number of Installs')
plt.show()
```

```python
categoty_type_installs = df.groupby(['Category','Type'])[['Installs']].sum().reset_index()
categoty_type_installs['log_Installs'] = np.log10(categoty_type_installs['Installs'])
     

plt.figure(figsize=(18,9))
plt.xticks(rotation=65,fontsize=9)
plt.xlabel("Category")
plt.ylabel("Installs")
plt.title("Number of installed Apps type wise according to Category")
sns.barplot(x = 'Category', y = 'log_Installs', hue='Type', data=categoty_type_installs)
```

```python
# Question 6: Top 5 Paid Apps with Highest Ratings and Installs
top_5_paid_apps = df[df['Type'] == 'Paid'].nlargest(5, 'Rating')
plt.figure(figsize=(12, 6))
sns.barplot(x='Rating', y='App Name', data=top_5_paid_apps, palette='CMRmap')
plt.title('Top 5 Paid Apps with Highest Ratings')
plt.xlabel('Rating')
plt.ylabel('App Name')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
Teen = df[df['Content Rating'] == 'Teen']

# Group by 'Category' and find the maximum installs for each category
Teen_grouped = Teen.groupby(['Category'])['Maximum Installs'].max().sort_values(ascending=False)
Teen_top_10 = Teen_grouped.head(10)

# Plot the donut chart
plt.pie(x=Teen_top_10, autopct="%.1f%%", labels=Teen_top_10.index, wedgeprops=dict(width=0.4))
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Teen Installing Apps in Terms of Category", fontsize=14)
plt.show()
```
